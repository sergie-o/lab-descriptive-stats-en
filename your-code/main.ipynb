#### Show the frequency distribution of the amount spent in Spain.

```python
# Filter orders for Spain
spain_orders = Orders[Orders["Country"] == "Spain"]

# Plot histogram of amount spent in Spain
plt.hist(spain_orders["amount_spent"], bins=10, color='blue', edgecolor='black')
plt.title("Frequency of Amount Spent in Spain")
plt.xlabel("Amount Spent")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(Orders.loc[Orders["Country"] == "Spain", "amount_spent"], 
         bins=10, color='blue', edgecolor='black')
plt.title("Frequency of Amount Spent in Spain")
plt.xlabel("Amount Spent")
plt.ylabel("Frequency")
plt.show()
```

#### Select all orders made in the month of August

```python
plt.hist(
    Orders.loc[Orders["month"] == 8, "Quantity"],
    bins=10,
    color='blue',
    edgecolor='black'
)
plt.title("Frequency of Quantity Ordered in August")
plt.xlabel("Quantity Ordered")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(
    Orders.loc[Orders["month"] == 8, "amount_spent"],
    bins=10,
    color='blue',
    edgecolor='black'
)
plt.title("Frequency of Amount Spent in August")
plt.xlabel("Amount Spent")
plt.ylabel("Frequency")
plt.show()
```
